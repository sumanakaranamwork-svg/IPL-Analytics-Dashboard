```python
print("Number of Seasons:", matches["season"].nunique())
print("Seasons:", sorted(matches["season"].unique()))
```

```python
matches_per_season = matches.groupby("season").size()

print(matches_per_season)
```

```python
matches_per_season.plot(kind="bar", figsize=(10,5))

plt.title("Matches Played per Season")
plt.xlabel("Season")
plt.ylabel("Number of Matches")

plt.show()
```

```python
team_wins = matches["winner"].value_counts()

team_wins
```

```python
team_wins.plot(kind ="bar",figsize = (12,5))
plt.title("Most Successful IPL Teams")
plt.xlabel("Team")
plt.ylabel("Matches Won")

plt.xticks(rotation= 90)

plt.show()
```

```python
top_runs = deliveries.groupby("batter")["batsman_runs"].sum().sort_values(ascending=False)

top_runs.head(10)
```

```python
top_runs.head(10).plot(kind="bar", figsize=(10,5))

plt.title("Top 10 Run Scorers")
plt.xlabel("Batter")
plt.ylabel("Runs")

plt.xticks(rotation=45)

plt.show()
```

```python
wickets = deliveries[deliveries["is_wicket"] == 1]

top_bowlers = wickets.groupby("bowler").size().sort_values(ascending=False)

top_bowlers.head(10)
```

```python
top_bowlers.head(10).plot(kind="bar", figsize=(10,5))

plt.title("Top 10 Wicket Takers")
plt.xlabel("Bowler")
plt.ylabel("Wickets")

plt.xticks(rotation=45)

plt.show()
```

```python
sixes = deliveries[deliveries["batsman_runs"] == 6]

top_six_hitters = sixes.groupby("batter").size().sort_values(ascending=False)

top_six_hitters.head(10)
```

```python
top_six_hitters.head(10).plot(kind="bar", figsize=(10,5))

plt.title("Top Six Hitters")
plt.xlabel("Player")
plt.ylabel("Sixes")

plt.xticks(rotation=45)

plt.show()
```
